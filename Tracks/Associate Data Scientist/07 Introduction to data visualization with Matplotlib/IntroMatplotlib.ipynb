## Using the matplotlib.pyplot interface

There are many ways to use Matplotlib. Here, we will focus on the `pyplot` interface, which provides the most flexibility in creating and customizing data visualizations.

Initially, we will use the pyplot interface to create two kinds of objects: `Figure` objects and `Axes` objects.

```python
# Import the matplotlib.pyplot submodule and name it plt
import matplotlib.pyplot as plt

# Create a Figure and an Axes with plt.subplots
fig, ax = plt.subplots()

# Call the show function to show the result
plt.show()
```

```python
# read and prep the data
import pandas as pd

df_seattle = pd.read_csv('data/seattle_weather.csv') # MONTH column was manually added to this dataset
seattle_locations = set(df_seattle['NAME'])
print(seattle_locations)
print()
print(df_seattle.columns)

df_seattle.head()
```

```python
df_austin = pd.read_csv('data/austin_weather.csv')   # MONTH renamed to MONTH_NUM and new MONTH columns added
print(df_austin.columns)
df_austin.head()
```

```python
# grab subsets we want to plot
seattle_weather = df_seattle.loc[:, ["NAME", "MONTH", "MLY-PRCP-NORMAL", "MLY-PRCP-25PCTL", "MLY-PRCP-75PCTL",
                                     "MLY-TAVG-NORMAL", "MLY-TAVG-STDDEV"]]
print(f"row count of original seattle_weather df: {seattle_weather.shape[0]}")
# There are 3 locations that have "SEATTLE" in them. Use the first of the 3 listed below:
# SEATTLE TACOMA INTERNATIONAL AIRPORT, WA US | SEATTLE BOEING FIELD, WA US | SEATTLE SAND PT WSFO, WA US
seattle_weather = seattle_weather.loc[seattle_weather["NAME"].str.contains("SEATTLE TACOMA"), :]
print(f"row count of SEATTLE TACOMA... subset: {seattle_weather.shape[0]}")
# austin_weather = df_austin.loc[:, ["MONTH", "MLY-PRCP-NORMAL"]]
```

```python
# Import the matplotlib.pyplot submodule and name it plt
import matplotlib.pyplot as plt

# Create a Figure and an Axes with plt.subplots
fig, ax = plt.subplots()
# Plot MLY-PRCP-NORMAL from seattle_weather against the MONTH
ax.plot(seattle_weather["MONTH"], seattle_weather["MLY-PRCP-NORMAL"],
        marker='o', color='b', linestyle='--')

# Plot MLY-PRCP-NORMAL from austin_weather against MONTH
austin_weather = df_austin.loc[:, ["NAME", "MONTH", "MLY-PRCP-NORMAL", "MLY-PRCP-25PCTL", "MLY-PRCP-75PCTL",
                                   "MLY-TAVG-NORMAL", "MLY-TAVG-STDDEV"]]
ax.plot(austin_weather["MONTH"], austin_weather["MLY-PRCP-NORMAL"],
        marker='v', color='r', linestyle='--')

# Call the show function
plt.show()  # should looks like ex01.20.png
```

## Customizing data appearance

We can customize the appearance of data in our plots, while adding the data to the plot, using key-word arguments to the plot command.

Here, we'll customize the appearance of the markers, the linestyle that is used, and the color of the lines and markers for the data.

We'll continue to use the DataFrame objects: seattle_weather and austin_weather. These each have a "MONTHS" column and a "MLY-PRCP-NORMAL" that we'll plot against each other.

```python
# rercreate Figure and Axes
fig, ax = plt.subplots()

# Plot Seattle data, setting data appearance
ax.plot(seattle_weather["MONTH"], seattle_weather["MLY-PRCP-NORMAL"], marker='o', color='blue')

# Plot Austin data, setting data appearance
ax.plot(austin_weather["MONTH"], austin_weather["MLY-PRCP-NORMAL"], marker='v', color='red')

# Call show to display the resulting plot
plt.show()
```

## Customizing axis labels and adding titles

Here, we'll customize the axis labels using the `set_xlabel` and `set_ylabel` methods of the Axes object. We'll then add a title uses the `set_title` method.

```python
# rercreate Figure and Axes
fig, ax = plt.subplots()

ax.plot(seattle_weather["MONTH"], seattle_weather["MLY-PRCP-NORMAL"])
ax.plot(austin_weather["MONTH"], austin_weather["MLY-PRCP-NORMAL"])

# Customize the x-axis label
ax.set_xlabel("Time (months)")

# Customize the y-axis label
ax.set_ylabel("Precipitation (inches)")

# Add the title
ax.set_title("Weather patterns in Austin and Seattle")

# Display the figure
plt.show()
```

## Creating small multiples with plt.subplots

Small multiples are used to plot several datasets side-by-side. In Matplotlib, small multiples can be created using the `plt.subplots()` function. The first argument is the number of **rows** in the array of Axes objects generate and the second argument is the number of **columns**. Here, we'll use the Austin and Seattle data to create and populate an array of subplots.

We'll cont. to use the DataFrames: `seattle_weather` and `austin_weather`. These each have a "MONTH" column and "MLY-PRCP-NORMAL" (for average precipitation), as well as "MLY-TAVG-NORMAL" (for average temperature) columns. Here, we'll plot in a separate subplot the monthly average precipitation and average temperatures in each city.

```python
# Create a Figure and an array of subplots with 2 rows and 2 columns
fig, ax = plt.subplots(2, 2)

# Addressing the top left Axes as index 0, 0, plot month and Seattle precipitation
ax[0, 0].plot(seattle_weather["MONTH"], seattle_weather["MLY-PRCP-NORMAL"])

# In the top right (index 0,1), plot month and Seattle temperatures
ax[0, 1].plot(seattle_weather["MONTH"], seattle_weather["MLY-TAVG-NORMAL"])

# In the bottom left (1, 0) plot month and Austin precipitations
ax[1, 0].plot(austin_weather["MONTH"], austin_weather["MLY-PRCP-NORMAL"])

# In the bottom right (1, 1) plot month and Austin temperatures
ax[1, 1].plot(austin_weather["MONTH"], austin_weather["MLY-TAVG-NORMAL"])
plt.show()
```

## Small multiples with shared y axis

When creating small multiples, it is often preferable to make sure that the different plots are displayed with the same scale used on the y-axis. This can be configured by setting the `sharey` key-word to `True`.

Here, we'll create a `Figure` with two `Axes` objects that share their y-axis. As before, the data is provided in seattle_weather and austin_weather DataFrames.

```python
# Create a figure and an array of axes: 2 rows, 1 column with shared y axis
fig, ax = plt.subplots(2, 1, sharey=True)

# Plot Seattle precipitation data in the top axes
ax[0].plot(seattle_weather['MONTH'], seattle_weather["MLY-PRCP-NORMAL"], color = 'b')
ax[0].plot(seattle_weather['MONTH'], seattle_weather["MLY-PRCP-25PCTL"], color = 'b', linestyle = '--')
ax[0].plot(seattle_weather['MONTH'], seattle_weather["MLY-PRCP-75PCTL"], color = 'b', linestyle = '--')

# Plot Austin precipitation data in the bottom axes
ax[1].plot(austin_weather['MONTH'], austin_weather["MLY-PRCP-NORMAL"], color = 'r')
ax[1].plot(austin_weather['MONTH'], austin_weather["MLY-PRCP-25PCTL"], color = 'r', linestyle = '--')
ax[1].plot(austin_weather['MONTH'], austin_weather["MLY-PRCP-75PCTL"], color = 'r', linestyle = '--')

plt.show()
```

## Plotting time-series with different variables

Add a second y-axis and scale it so it can be more easily compared. Start by plotting 2 different variables using the same y-axis:

```python
import matplotlib.pyplot as plt 
fig, ax = plt.subplots()
# without scaling
ax.plot(df_climate_change.index, df_climate_change["co2"]) 
ax.plot(df_climate_change.index, df_climate_change["relative_temp"]) 
ax.set_xlabel('Time') 
ax.set_ylabel('CO2 (ppm) / Relative temperature') 
plt.show() 
```

```python
fig, ax = plt.subplots()
# create a 2nd y-axis and scale it
ax.plot(df_climate_change.index, df_climate_change["co2"], color='blue')
ax.set_xlabel('Time') 
ax.set_ylabel('CO2 (ppm)', color='blue')

ax.tick_params('y', colors='blue')  # make 1st (left) y-axis labels blue

ax2 = ax.twinx()  # share the x-axis (but create new y-axis)
ax2.plot(df_climate_change.index, df_climate_change["relative_temp"], color = 'red') 

ax2.tick_params('y', colors='red')  # make 2nd (right) y-axis labels red

ax2.set_ylabel('Relative temperature (Celsius)', color='red') 
plt.show() 
```

```python
# encapsulate plot above as a function
def plot_timeseries(axes, x, y, color, xlabel, ylabel):
    axes.plot(x, y, color=color)
    axes.set_xlabel(xlabel)
    axes.set_ylabel(ylabel, color=color)
    axes.tick_params('y', colors=color)

```

```python
# test function
fig, ax = plt.subplots()
plot_timeseries(ax, df_climate_change.index, df_climate_change['co2'],
                'blue', 'Time', 'CO2 (ppm)')
ax2 = ax.twinx()
plot_timeseries(ax2, df_climate_change.index, df_climate_change['relative_temp'],
                'red', 'Time', 'Relative temperature (Celsius)')

plt.show()
```

## Historgrams

Start by creating a dataframe of medal winners in mens gymnastics and mens rowing. Looks like the lecture df sorted first on `Sport` and then on `ID`

```python
df_summer16 = pd.read_csv('data/summer2016.csv', index_col=0)
df_men16 = df_summer16.loc[(df_summer16['Sex'] == 'M') & \
                            ((df_summer16['Sport'] == 'Rowing') | \
                             (df_summer16['Sport'] == 'Gymnastics'))]
df_men16 = df_men16.sort_values(["Sport", "ID"], ascending=[False, True])
df_men16.head()  # looks like the lecture df
```

```python
df_men16_rowing = df_men16.loc[df_men16['Sport'] == 'Rowing']
df_men16_gymnastics = df_men16.loc[df_men16['Sport'] == 'Gymnastics']

# simple visual comparison of mean heights
fig, ax = plt.subplots()
ax.bar("Rowing", df_men16_rowing['Height'].mean())
ax.bar("Gymnasitics", df_men16_gymnastics['Height'].mean())
ax.set_ylabel("Height (cm)")
plt.show()
```

```python
# compare the distributions
fig, ax = plt.subplots()
ax.hist(df_men16_rowing['Height'], label="Rowing")
ax.hist(df_men16_gymnastics['Height'], label="Gymnastics")
ax.set_xlabel("Height (cm)")
ax.set_ylabel("# of observations")
ax.legend()
plt.show()
```

```python
# matplotlib uses 10 bins by default but can be customized with the bins arg
fig, ax = plt.subplots()
ax.hist(df_men16_rowing['Height'], label="Rowing", bins=5)
ax.hist(df_men16_gymnastics['Height'], label="Gymnastics", bins=5)
ax.set_xlabel("Height (cm)")
ax.set_ylabel("# of observations")
ax.legend()
plt.show()
```

```python
# we can pass the boundaries of custom set of bins
my_bins = [150 + (x * 10) for x in range(7)]  # [150, 160, 170, 180, 190, 200, 210] shown manually created in video

fig, ax = plt.subplots()
ax.hist(df_men16_rowing['Height'], label="Rowing", bins=my_bins)
ax.hist(df_men16_gymnastics['Height'], label="Gymnastics", bins=my_bins)
ax.set_xlabel("Height (cm)")
ax.set_ylabel("# of observations")
ax.legend()
plt.show()
```

```python
# eliminate occlusion by changing transparency
fig, ax = plt.subplots()
ax.hist(df_men16_rowing['Height'], label="Rowing", bins=my_bins, histtype="step")
ax.hist(df_men16_gymnastics['Height'], label="Gymnastics", bins=my_bins, histtype="step")
ax.set_xlabel("Height (cm)")
ax.set_ylabel("# of observations")
ax.legend()
plt.show()
```

## Statistical plotting

+ Adding error bars to bar plots
+ Adding error bars to line plots
+ Creating boxplots

```python
# add error bars to the bar chart
fig, ax = plt.subplots()

ax.bar("Rowing",
       df_men16_rowing["Height"].mean(),
       yerr=df_men16_rowing["Height"].std())

ax.bar("Gymnastics",
       df_men16_gymnastics["Height"].mean(),
       yerr=df_men16_gymnastics["Height"].std())

ax.set_ylabel("Height (cm)")
plt.show()
```

```python
# add error bars to line plots
fig, ax = plt.subplots()

ax.errorbar(x=seattle_weather['MONTH'],
            y=seattle_weather['MLY-TAVG-NORMAL'],
            yerr=seattle_weather['MLY-TAVG-STDDEV'])

ax.errorbar(x=austin_weather['MONTH'],
            y=austin_weather['MLY-TAVG-NORMAL'],
            yerr=austin_weather['MLY-TAVG-STDDEV'])

ax.set_ylabel("Temperature (Fahrenheit)")

plt.show()
```

```python
# create boxplots
fig, ax = plt.subplots()
ax.boxplot([df_men16_rowing['Height'], df_men16_gymnastics['Height']])
ax.set_xticklabels(["Rowing", "Gymnastics"])
ax.set_ylabel("Height (cm)")
plt.show()
```

## Quantitative comparisons: scatter plots

+ Use color for comparison

```python
# bivariate comparison
fig, ax = plt.subplots()
ax.scatter(df_climate_change["co2"], df_climate_change["relative_temp"])
ax.set_xlabel("CO2 (ppm)")
ax.set_ylabel("Relative temperature (Celcius)")
plt.show()
```

## Saving your visualizations

To save a figure, we just need to replace the call to `plt.show()` with `fig.savefig(path_to_file)`. Note, when rendering locally, we call the `show()` method on the **plt** object. When exporting/saving a visualization, we call the `savefig()` method on the **fig** object.
